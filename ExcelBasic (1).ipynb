1. Explain the difference between Relative, Absolute and Mixed Cell Referencing.

   You have the following dataset:

   | Vehicle       | Price  |
   | Car           | 500000 |
   | Bicycle       | 10500  |
   | Auto-Rickshaw | 25000  |
   - Cell referencing in Excel is used to refer to a cell or group of cells while creating formulas. It helps perform calculations efficiently. There are three types of cell references.

     Relative Cell Referencing

     Relative referencing is the default reference in Excel. When a formula is copied from one cell to another, the cell reference automatically changes according to the new position.

     Example formula:
     =B2*1.10
     If the formula is copied to the next row, it becomes:
     =B3*1.10
     Here the reference changes automatically, so it is called Relative Referencing

     Absolute Cell Referencing

     Absolute referencing does not change when the formula is copied. It uses the dollar ($) symbol to fix both the row and column.

     Example:
     =B2*$D$1
     In this case, $D$1 remains fixed even if the formula is copied to other cells.

     Mixed Cell Referencing

     Mixed referencing means either the row or the column is fixed.

     Examples:
     $A1   (Column fixed, row changes)
     A$1   (Row fixed, column changes)

     This type of reference is useful when copying formulas across rows or columns in a table.

 3. Using the data below, write a formula to calculate the average sales       of       items priced above 100 but less than 300 ?

    | Item   | Price | Sales |
    | Item A | 90    | 1000  |
    | Item B | 150   | 1200  |
    | Item C | 250   | 1500  |
    - To calculate the average sales of items priced above 100 but less than   300, the AVERAGEIFS function is used.

      Formula:

      = AVERAGEIFS(E2:E51,F2:F51,">100",F2:F51,"<300")

      In the given dataset, there are no items with unit price greater than 100 and less than 300.
      Therefore, the formula returns no value (or 0) because no records satisfy the condition.

4. Count how many customer names are recorded ?

   - To count the total number of customer names recorded in the dataset,  we    use the **COUNTA** function in Excel. The COUNTA function counts   all  non-empty cells in a given range.

     Assume that customer names are stored in column B from cell B2 to B51.

     The formula to count the number of customer names is:

     = COUNTA(B2:B51)

     This formula counts all the non-empty cells in the specified range and gives the total number of customer names recorded.

     Therefore, the total number of customer names recorded is 50.

     The formula for calculating Total Sales in the first row (for example, row 2) will be:

     = E2 * F2

     This formula multiplies the quantity sold by the price per unit for each row and gives the total sales value.

     Therefore, Total Sales for each row is calculated using the multiplication formula:

     = Quantity × Price per Unit

6. Calculate the total sales of Notebooks in the North region only ?
   - In the given dataset:

     Column C represents Region

     Column D represents Product

     Column E represents Quantity

     Column F represents Price per Unit

     First, Total Sales for each row is calculated as:

     Total Sales = Quantity × Price per Unit

     To calculate the total sales of Notebooks in the North region only, we use the SUMIFS function because we have multiple conditions.

     The formula is:

     = SUMIFS(E31:E51*F31:F51, C31:C51, "North", D31:D51, "Notebook")

     However, since Excel does not directly multiply ranges inside SUMIFS, it is better to first create a Total Sales column (for example, Column H) using:

     = E31 * F31

     Then apply the following formula:

     = SUMIFS(H31:H51, C31:C51, "North", D31:D51, "Notebook")

      This formula adds the total sales only where:

      Region is "North"

      Product is "Notebook"

     Therefore, this calculates the total sales of Notebooks in the North region only.
